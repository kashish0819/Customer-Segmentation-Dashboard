```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("data/customer_segmentation.csv")

df.head()
```

```python
df.describe().T
```

```python
df.hist(
    figsize=(14,10),
    bins=20
)

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.histplot(
    df["Age"],
    bins=20,
    kde=True
)

plt.title("Age Distribution")
plt.show()
#Majority customers belong to young and middle-age groups.
```

```python
plt.figure(figsize=(6,4))

sns.countplot(
    x="Gender",
    data=df
)

plt.title("Gender Distribution")
plt.show()
#Customer base is balanced across genders.
```

```python
plt.figure(figsize=(8,5))

sns.histplot(
    df["Income Level"],
    bins=20,
    kde=True
)

plt.title("Income Level Distribution")
plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.histplot(
    df["Coverage Amount"],
    bins=20,
    kde=True
)

plt.title("Coverage Amount Distribution")
plt.show()
```

```python
numeric_cols = df.select_dtypes(include=np.number)

for col in numeric_cols:

    plt.figure(figsize=(8,4))

    sns.boxplot(
        x=df[col]
    )

    plt.title(col)

    plt.show()
```

```python
plt.figure(figsize=(10,6))

sns.heatmap(
    df.corr(numeric_only=True),
    annot=True,
    cmap="coolwarm"
)

plt.title("Correlation Matrix")

plt.show()
```

```python
plt.figure(figsize=(10,6))

sns.scatterplot(
    data=df,
    x="Income Level",
    y="Coverage Amount"
)

plt.title("IncomeLevel vs Coverage Amount")

plt.show()
#Customers naturally form spending groups.

#This validates clustering.
```

```python
plt.figure(figsize=(10,6))

sns.scatterplot(
    data=df,
    x="Age",
    y="Coverage Amount"
)

plt.title("Age vs Spending")

plt.show()
```

```python
sns.pairplot(
    df.select_dtypes(include=np.number)
)

plt.show()
```

```python
#1. coverage amount behavior differs significantly.

#2. Income and coverage amount create natural customer groups.

#3. Outliers are minimal.

#4. Dataset is highly suitable for customer segmentation.
```
