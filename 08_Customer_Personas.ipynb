```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(
    "data/clustered_customers.csv"
)

df.head()
```

```python
cluster_summary = df.groupby(
    "Cluster"
)[
    [
        "Age",
        "Income Level",
        "Coverage Amount",
        "Premium Amount"
    ]
].mean()

cluster_summary.round(2)
```
