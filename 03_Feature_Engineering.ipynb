```python
import pandas as pd

df = pd.read_csv("data/customer_segmentation.csv")
```

```python
df.isnull().sum()
```

```python
df = df.drop_duplicates()
```

```python
df["Age_Group"] = pd.cut(
    df["Age"],
    bins=[0,25,40,60,100],
    labels=[
        "Young",
        "Adult",
        "Middle Age",
        "Senior"
    ]
)

df.head()
```

```python
df["Income_Level"] = pd.qcut(
    df["Income Level"],
    q=4,
    labels=[
        "Low",
        "Medium",
        "High",
        "Premium"
    ]
)
```

```python
df["Spending_Category"] = pd.qcut(
    df["Coverage Amount"],
    q=3,
    labels=[
        "Low",
        "Medium",
        "High"
    ]
)
```

```python
df["Gender"] = df["Gender"].map(
    {
        "Male":1,
        "Female":0
    }
)
```

```python
df.to_csv(
    "data/cleaned_customers.csv",
    index=False
)

print("Saved Successfully")
```
